การแสดงผลบนหน้าจอที่ผ่านมาจะเป็นสีขาวดำ แต่เราสามารถแสดงผลให้ออกมามีสีได้โดยการใส่รหัส ANSI Escape Sequence ของสีที่ต้องการแสดงได้ โดยรหัสใช้รหัส <font color=blue>`'\033['`</font> ตามด้วยรหัสสีที่เป็น foreground และ background แล้วจบด้วย <font color=blue>`'m'`</font> โดยรหัสสีเป็นค่าจำนวนเต็มที่แสดงในตารางข้างล่าง

| Color	| Foreground Code | Background Code |
| --- | :-: | :-: |
| Black | 30 | 40
| Red | 31 | 41
| Green | 32 | 42
| Yellow | 33 | 43
| Blue | 34 | 44
| Magenta | 35 | 45
| Cyan | 36	| 46
| White | 37 | 47
| highlight | | 103

รหัส escape sequence ที่จะใช้ จะทำแบบง่ายๆ คือ ถ้าต้องการแสดงข้อความเป็น <font color=red>สีแดง</font> ก็จะเอา escape sequence ของสีแดง มาวางหน้าข้อความ แล้วก็เอารหัส reset `'\033[0m'` วางไวัหลังข้อความ

color | escape sequence
--- | ---
red | `\033[;31m`
green | `\033[;32m`
blue | `\033[;34m`
highlight | `\033[;103m`
reset | `\033[0m`

ถ้าต้องการให้ข้อความ 'Python' แสดงเป็นสีน้ำเงิน ก็เอารหัสสีน้ำเงินวางหน้าข้อความและรหัส reset วางหลังข้อความ เป็น `'\033[;34mPython\033[0m'` นิสิตสามารถทดลองได้โดยเอาคำสั่ง

`print('\033[;34mPython\033[0m')` ไปลอง run ดูแล้วลองเปลี่ยนเลขหน้า `m` เป็นค่าอื่นๆ ดูได้

2. **`color_text(s, color)`** คืนค่าเป็นสตริงที่ครอบด้วยรหัสสีและรหัส reset ตามชื่อสีตามที่ระบุใน `color` โดยจะมีแค่ 3 สีคือ `'red'`, `'green'` และ `'blue'` โดยชื่อสีสามารถเขียนได้ทั้งตัวเล็กหรือตัวใหญ่ ถ้า `color` เป็นค่าอื่นๆ ก็จะคืนเป็นสตริง s ที่ไม่มีสี 

 *ตัวอย่าง*

คำสั่ง | ผลที่แสดงทางหน้าจอ

`print(color_text('This is red text', 'red')` | <font color=red>`This is red text`</font>
`print(color_text('This is green text', 'GreeN')` | <font color=green>`This is green text`</font>
`print(color_text('This is blue text', 'BLUE')` | <font color=blue>`This is blue text`</font>
`print(color_text('This is yellow text', 'yellow')` | `This is yellow text`

---
3. **`color_tag(s)`** คืนค่าเป็นสตริงที่แสดงผลเป็นสีตาม tag ที่ครอบข้อความไว้ โดยข้อความ s จะขึ้นต้นด้วย tag `<red>, <green>, หรือ <blue>` และจบด้วย `</>` เสมอ คำว่า `red`, `green`, `blue` จะเป็นตัวพิมพ์เล็กเท่านั้น และ tag ที่ใส่มาจะมีแค่ `<red`>, `<green>`, `<blue>` เท่านั้น <font color=red>และจะไม่มี tag ที่เขียนซ้อนกัน</font> เช่น `<red>ABC <blue>DEF</> XYZ</>`

 *ตัวอย่าง*

คำสั่ง | ผลที่แสดงทางหน้าจอ
--- | ---
`print(color_tag('<red>red text</>')` | <font  color=red>`red text`</font>
`print(color_tag('<green>green text</>')` | <font color=green>`green text`</font>
`print(color_tag('<blue>blue text</>')` | <font color=blue>`blue text`</font>

---

```python
# HW4_String_File_Processing (ไม่ลบหรือแก้ไขบรรทัดนี้ หรือเพิ่มอะไรก่อนบรรทัดนี้ โดยเด็ดขาด)

# - เขียนในเซลล์นี้เท่านั้น 
# - ถ้าต้องการเขียนฟังก์ชันเพิ่ม ก็เขียนในเซลล์นี้

# ตัวแปรต่อไปนี้เป็นตัวแปรที่กำหนดค่าสีเพื่อเอาไว้ใช้ในโปรแกรม (ห้ามแก้ไข)
RESET = '\033[0m'
RED = '\033[;31m'
GREEN = '\033[;32m'
BLUE = '\033[;34m'
HIGHLIGHT = '\033[;103m'

# ---------------------------------------------------
def red(s: str):
    return RED+s+RESET
# ---------------------------------------------------
def green(s: str):
    return GREEN+s+RESET
# ---------------------------------------------------
def blue(s: str):
    return BLUE+s+RESET
# ---------------------------------------------------
def highlight(s: str):
    return HIGHLIGHT+s+RESET
# ---------------------------------------------------
def color_text(s: str, c: str):
    if c.lower() == 'blue':
        return blue(s)
    elif c.lower() == 'red':
        return red(s)
    elif c.lower() == 'green':
        return green(s)
    else:
        return s
# ---------------------------------------------------
def color_tag(s: str):
    return s.replace('</>', RESET).replace('<red>', RED).replace('<green>', GREEN).replace('<blue>', BLUE)
# ---------------------------------------------------
def highlight_words(s: str, words: list[str]):
    for e in words:
        where = s.lower().find(e.lower())
        while where != -1:
            s = s[:where] + highlight(s[where:where+len(e)]) + s[where+len(e):]
            where = s.lower().find(e.lower(),where+len(highlight(s[where:where+len(e)])))
    return s
# ---------------------------------------------------
def display_tag_file(filename: str):
    with open(filename) as f:
        print(color_tag(f.read()))
```
