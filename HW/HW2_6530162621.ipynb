 - การทำงานในฟังก์ชัน `get_sum` ต้องใช้เวลาไม่นาน (เช่น `data` มีข้อมูลสักหนึ่งแสนช่อง ก็ไม่น่าใช้เวลาเกิน 1 นาที)

```python
# HW02 (ไม่ลบหรือแก้ไขบรรทัดนี้ หรือเพิ่มอะไรก่อนบรรทัดนี้ โดยเด็ดขาด)

def get_sum(data):
    students = []
    score = []

    for i in range(0,len(data),2):
        if data[i] not in students:
            students.append(data[i])
            score.append([data[i+1]])
        else:
            score[students.index(data[i])].append(data[i+1])

    ans = []
    for i in range(len(students)):
        ans.append(students[i])
        if len(score[i]) > 3:
            score[i].sort()
            del score[i][0]
        ans.append(sum(score[i]))

    return ans

def get_sum2(data):
    ans  = []
    for i in range(0, len(data), 2):
        if len(ans) != 0 and ans[-2] == data[i]:
            ans[-1].append(data[i+1])
        else:
            ans.append(data[i])
            ans.append([data[i+1]])

    for i in range(1, len(ans), 2):
        if len(ans[i]) > 3:
            ans[i].sort()
            ans[i][0] = 0
        ans[i] = sum(ans[i])

    return ans


# -----------------------------
d = [
  6610013121,4,
  6610021021,5, 6610021021,5,
  6610061121,5, 6610061121,5, 6610061121,1,
  6610000121,3, 6610000121,2, 6610000121,2, 6610000121,3
]


out = get_sum2(d)
print(out) # ต้องแสดง [6610013121,4, 6610021021,10, 6610061121,11, 6610000121,8]
# ข้อแนะนำ: ควรทดสอบ get_sum ด้วยข้อมูลชุดอื่นด้วย (ลิสต์ d ข้างบนอาจมีไม่ครอบคลุมทุกกรณี)
```
